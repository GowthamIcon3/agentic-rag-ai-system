Now, import the required classes for CrewAI and initialize the `ChatGoogleGenerativeAI` LLM.

```python
from crewai import Agent, Task, Crew
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",  # You can also try gemini-1.5-pro
    temperature=0.7
)
```

Define your agents and tasks for the CrewAI project.

```python
# Research Agent
research_agent = Agent(
    role="Senior Research Analyst",
    goal="Research deeply about the given topic",
    backstory="You are an expert research analyst with 10 years of experience.",
    verbose=True,
    llm=llm
)

# Content Writer Agent
writer_agent = Agent(
    role="Content Writer",
    goal="Create engaging content based on research",
    backstory="You are a professional content writer specialized in blogs and reports.",
    verbose=True,
    llm=llm
)
```

```python
research_task = Task(
    description="Research about 'AI Agent Engineering' and provide key insights.",
    expected_output="Detailed research report with bullet points.",
    agent=research_agent
)

writing_task = Task(
    description="Write a blog article based on the research provided.",
    expected_output="Well-structured blog article with headings and conclusion.",
    agent=writer_agent
)
```

Finally, create and kickoff the CrewAI instance to run the agents and tasks.

```python
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[research_task, writing_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

```python
import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
os.environ["TAVILY_API_KEY"] = userdata.get("TAVILY_API_KEY")
```

```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.5
)
```

```python
documents = [
    "AI Agent Engineering involves building autonomous systems that plan and execute tasks.",
    "RAG improves LLM accuracy by retrieving relevant documents before generating answers.",
    "CrewAI enables orchestration of multiple specialized AI agents."
]
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vectorstore = Chroma.from_texts(
    documents,
    embedding=embeddings,
    persist_directory="./rag_db"
)

retriever = vectorstore.as_retriever()
```

**CREATE AGENTS**

```python
planner = Agent(
    role="AI Project Planner",
    goal="Plan research steps for the given topic",
    backstory="Expert AI strategist",
    llm=llm,
    verbose=True
)
```

```python
from crewai_tools import TavilySearchTool
```

```python
tavily_tool = TavilySearchTool(
    search_depth="advanced"  # basic or advanced
)
```

```python
from crewai import Agent

research_agent = Agent(
    role="Senior Research Analyst",
    goal="Search the web and gather the latest insights",
    backstory="Expert at extracting reliable information from the internet",
    tools=[tavily_tool],   # 🔥 TOOL ADDED HERE
    llm=llm,
    verbose=True
)
```

```python
rag_agent = Agent(
    role="Knowledge Retrieval Specialist",
    goal="Retrieve relevant internal knowledge for the topic",
    backstory="Expert at extracting data from vector databases",
    llm=llm,
    verbose=True
)
```

```python
writer = Agent(
    role="Technical Content Writer",
    goal="Write a structured article using research and retrieved knowledge",
    backstory="Expert AI writer",
    llm=llm,
    verbose=True
)
```

**#CREATE TASKS**

```python
plan_task = Task(
    description="Create a structured plan to research AI Agent Engineering.",
    expected_output="A structured research plan outlining key areas to investigate for AI Agent Engineering.",
    agent=planner
)

research_task = Task(
    description="Search the web and gather latest information about AI Agent Engineering.",
    expected_output="A detailed report summarizing the latest information and trends in AI Agent Engineering from web research.",
    agent=research_agent
)

rag_task = Task(
    description="Retrieve relevant internal knowledge about AI Agent Engineering from vector database.",
    expected_output="A collection of internal knowledge and facts about AI Agent Engineering retrieved from the vector database.",
    agent=rag_agent
)

write_task = Task(
    description="Write a detailed blog article combining web research and retrieved knowledge.",
    expected_output="A comprehensive and well-structured blog article on AI Agent Engineering, incorporating insights from both web research and internal knowledge.",
    agent=writer
)
```

**CREATE CREW AND RUN**

```python
from crewai import Crew

crew = Crew(
    agents=[rag_agent],
    tasks=[rag_task],
    verbose=True
)

result = crew.kickoff()

print("\nFINAL OUTPUT:\n")
print(result)
```
